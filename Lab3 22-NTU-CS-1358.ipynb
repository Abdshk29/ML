***Generate Classes***

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Generate numeric data
n_samples = 3500
n_classes = 7
centers = np.array([[-2, -2], [-2, 1], [2, -2], [2, 2], [0, 0], [4, 0], [0, 4]])
X, y = make_blobs(n_samples=n_samples, centers=centers,)
data = pd.DataFrame(X, columns=['x1', 'x2'])
data['target'] = y
print(data.head())
```

***Plot Data***

```python
# Plot the data points by classes
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'black']
for cls in data['target'].unique():
    plt.scatter(data[data['target'] == cls]['x1'], data[data['target'] == cls]['x2'], color=colors[cls], label=f'class {cls}')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data points')
plt.legend()
plt.show() 
# Save data to a CSV file
data.to_csv('Generated data.csv', index=False)
# Load data from saved csv file
data = pd.read_csv('Generated data.csv')
print("Data loaded into data frame")
print(data.head())
```

***Applying KNN to Synthetic Data***

```python
# KNN Classification with different values of k
X = data[['x1', 'x2']]
y = data['target']
k_values = [2,17,99,165,220,300,375]
for k in k_values:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy = accuracy_score(y_test, knn.predict(X_test))
    print(f'Accuracy for k={k}: {accuracy:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Generate numeric data
n_samples = 5000
n_classes = 5
centers = np.array([[1, 0], [-2, 0], [-1, 1], [-2, -2], [0, 1], [1, 1], [0, 2]])
X, y = make_blobs(n_samples=n_samples, centers=centers)
data = pd.DataFrame(X, columns=['x1', 'x2'])
data['target'] = y
print(data.head())
```

```python
# Plot the data points by classes
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'black']
for cls in data['target'].unique():
    plt.scatter(data[data['target'] == cls]['x1'], data[data['target'] == cls]['x2'], color=colors[cls], label=f'class {cls}')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data points')
plt.legend()
plt.show()
# Save the data to a CSV file
data.to_csv('Activity1 data.csv', index=False)
# Load data from saved csv file
data = pd.read_csv('Activity1 data.csv')
print("Data loaded into data frame")
```

```python
# KNN Classification with different values of k
X = data[['x1', 'x2']]
y = data['target']
k_values = [8, 27, , 165, 220, 300, 375]
for k in k_values:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy = accuracy_score(y_test, knn.predict(X_test))
    print(f'Accuracy for k={k}: {accuracy:.2f}')
```
