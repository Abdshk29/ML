***Reducing features - PCA***

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
X = faces.data
pca = PCA()
pca.fit(X)
X_pca = pca.transform(X)
print("Transformed X shape:", X_pca.shape)
plt.figure(1, figsize=(12, 8))
plt.plot(pca.explained_variance_, linewidth=2)
plt.xlabel('Components')
plt.ylabel('Explained Variances')
plt.title('PCA Explained Variance')
plt.show()
```

```python
n_components = 90
pca = PCA(n_components=n_components, whiten=True)
pca.fit(X)
X_pca = pca.transform(X)
print(X_pca.shape)
print("Transformed X shape:", X_pca.shape)
plt.figure(1, figsize=(12, 8))
plt.plot(pca.explained_variance_, linewidth=2)
plt.xlabel('Components')
plt.ylabel('Explained Variances')
plt.show()
```

```python
print(pca.mean_.shape)
fig, ax = plt.subplots(1, 1, figsize=(8,8))
ax.imshow(pca.mean_.reshape((64, 64)), cmap="gray")
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Average Face')
```

```python
import pandas as pd
import numpy as np

# Define cities and their weather conditions
cities = {
    "Karachi": {"temp_mean": 26, "temp_var": 4, "humidity_mean": 70, "humidity_var": 10, "wind_mean": 15, "wind_var": 5},
    "Lahore": {"temp_mean": 24, "temp_var": 5, "humidity_mean": 60, "humidity_var": 12, "wind_mean": 12, "wind_var": 4},
    "Islamabad": {"temp_mean": 22, "temp_var": 6, "humidity_mean": 55, "humidity_var": 15, "wind_mean": 10, "wind_var": 3},
    "Quetta": {"temp_mean": 18, "temp_var": 7, "humidity_mean": 45, "humidity_var": 10, "wind_mean": 20, "wind_var": 6},
    "Peshawar": {"temp_mean": 23, "temp_var": 5, "humidity_mean": 50, "humidity_var": 14, "wind_mean": 14, "wind_var": 4},
}

# Generate random data
data = []
for city, params in cities.items():
    temp_values = np.random.normal(params["temp_mean"], params["temp_var"], 250)
    humidity_values = np.random.normal(params["humidity_mean"], params["humidity_var"], 250)
    wind_values = np.random.normal(params["wind_mean"], params["wind_var"], 250)
    
    city_data = pd.DataFrame({
        "Temperature": temp_values,
        "Humidity": humidity_values,
        "WindSpeed": wind_values,
        "City": city
    })
    
    data.append(city_data)

# Combine all data into a single DataFrame
df = pd.concat(data, ignore_index=True)

# Display first few rows
df.head()
df.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
n_classes = 4
data, labels = make_blobs(n_samples=1000,centers=n_classes)
labels[:7]
```

```python
import matplotlib.pyplot as plt
# Create a scatter plot
fig, ax = plt.subplots()
# Define colors for each cluster
colours = ['green', 'orange', 'blue', 'pink']
# Plot each cluster
for label in range(n_classes):
    ax.scatter(data[labels == label, 0],   # x-coordinates
               data[labels == label, 1],   # y-coordinates
               color=colours[label],        # point color
               s=40,                        # point size
               label=label)                 # label for the legend
```

```python
# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y’') 
ax.set_title('Blobs Examples')
 # Show legend 
ax.legend(loc='upper left') 
# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
centers = [[2, 3], [4, 5], [7, 9]]
data, labels = make_blobs(n_samples=1000,
centers=np.array(centers),
random_state=1)
print(labels[:7])
print(len(data))
print(data[:7])
```

```python
fig, ax = plt.subplots()
colours = ('green', 'orange', 'blue', 'pink')
n_classes = len(colours)  # Define the number of classes 
for label in range(n_classes):
    ax.scatter(
            x=data[labels==label, 0],
            y=data[labels==label, 1],
            c=colours[label],
            s=40,
            label=label)
ax.set(xlabel='X', ylabel='Y', title='Blobs Examples')
ax.legend(loc='upper left')
```
